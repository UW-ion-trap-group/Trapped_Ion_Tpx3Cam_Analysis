```python
fig,ax=plt.subplots(figsize=(5, 5))

n=6 #number of ions 

h2d=plt.hist2d(it['x'], it['y'], bins = (256, 256), range = [(0, 256), (0, 256)])
x_left,x_right,y_bottom,y_top=15,75,75,110 #estimated after visualizing h2d above
ax.set_xlim(x_left,x_right)    #left and right limits on plot
ax.set_ylim(y_bottom,y_top)    #top and bottom limits

grid_x=[20,31,40,46,53,60,68] #sub-section boundaries (1 more than ions)
ax.set_xticks(grid_x)
ax.grid(axis="x")

a=np.empty((n,2))                                           # empty  n by 2 matrix 
for i in range (1,n+1):
    
    temp=it[(it['x']>grid_x[i-1]) & (it['x'] <grid_x[i])]   # sub-dataframe for x values between 
                                                              #upper and lower bounds of box
    

    temp_2=temp.groupby(["x",'y']).size().idxmax()          #most frequently occuring x and y values
                                                             #in given "grid box"
    a[i-1][0]=temp_2[0]; a[i-1][1]=temp_2[1]                #append values to our array


x1=a[0][0]; y1=a[0][1]; x2=a[1][0]; y2=a[1][1]; x3=a[2][0]; y3=a[2][1]; 
x4=a[3][0]; y4=a[3][1]; x5=a[4][0]; y5=a[4][1]; x6=a[5][0]; y6=a[5][1]; 
print ("\nx:",[x1,x2,x3,x4,x5,x6],"\ny:",[y1,y2,y3,y4,y5,y6])

fig, (ax0) = plt.subplots(ncols=1, figsize=(2,6))
t=(it["time"]*25/(4096))                                   # in ns
h = ax0.hist2d(it['x'], t, bins = [256, 400])

ax0.set_xlabel('x position')
ax0.set_ylabel('Time (ns)')
ax0.set_title(filename)

#it['time'].max()
start = 0; end = t.max(); duration = end - start
ax0.set_ylim(start, end)
ax0.set_title(filename)
```
