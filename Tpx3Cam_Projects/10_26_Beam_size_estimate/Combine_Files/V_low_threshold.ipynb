```python
import re
import os 
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import glob
import time
import pandas as pd
plt.rcParams["figure.figsize"] = (3,3)

print (time.ctime())
```

```python
filename = 'P/x_389.5'
it = pd.read_csv(filename)
it.drop("Unnamed: 0",axis=1,inplace=True)
it.head(2)
```

```python
fig,ax=plt.subplots(figsize=(5, 5))
it = pd.read_csv(filename)
h2d=plt.hist2d(it['x'], it['y'], bins = (256, 256), range = [(0, 256), (0, 256)])
ax.set_xlim(25,75)
ax.set_ylim(80,110)

n=1
grid_x=np.linspace(25,75,2)
ax.set_xticks(grid_x)
ax.grid(axis="x")

a=np.empty((n,2))
for i in range (1,n+1):
    
    temp=it[(it['x']>grid_x[i-1]) & (it['x'] <grid_x[i])]                          # sub-dataframe for x values between upper and lower bounds of box
    temp_2=temp.groupby(["x",'y']).size().idxmax() 
    a[i-1][0]=temp_2[0]; a[i-1][1]=temp_2[1]

x1=a[0][0]; y1=a[0][1]
print (x1,y1)

fig, (ax0) = plt.subplots(ncols=1, figsize=(2,6))
t=(it["time"]*25/(4096))*1e-9
h = ax0.hist2d(it['x'], t, bins = [256, 400])
#ax0.set_xlim(45,49)
ax0.set_xlabel('x position')
ax0.set_ylabel('Time (s)')
ax0.set_title(filename)

#it['time'].max()
start = 0; end = t.max(); duration = end - start
ax0.set_ylim(start, end)
ax0.set_title(filename)
```

#### Combining Datasets 
File x_389.5_2 has been left out because I'm not sure if the ion melted. 

```python
filename = 'P/x_389.5'
it = pd.read_csv(filename)
print (len(it))

filename = 'P/x_389.5_1'
it_1 = pd.read_csv(filename)
print (len(it_1))

filename = 'P/x_389.5_2'
it_2 = pd.read_csv(filename)
print (len(it_2))
```

```python
it_1["time"]+=max(it["time"])
print ((len(it_1)))
```

```python
it_2["time"]+=max(it_1["time"])
```

```python
# Merging the three datasets with the assumption that data was taken in the order: it --> it_1 --> it_2
meta=pd.concat([it,it_1,it_2])
meta.drop("Unnamed: 0",axis=1,inplace=True)
meta.head(2)
```

```python
# checking elements to check merge
print (meta.iloc[len(it)],"\n",it_1.iloc[0])
```

```python
print (meta.iloc[len(it)+len(it_1)],"\n",it_2.iloc[0])
```

```python
meta.drop(columns="Unnamed: 0.1",inplace=True)
```
