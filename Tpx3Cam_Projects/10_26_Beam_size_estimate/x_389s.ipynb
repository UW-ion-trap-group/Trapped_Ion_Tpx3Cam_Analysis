```python
import re
import os 
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import glob
import time
import pandas as pd
plt.rcParams["figure.figsize"] = (3,3)

print (time.ctime())
```

```python
filename = 'x_389.5'
it = pd.read_csv(filename)
it.drop("Unnamed: 0",axis=1,inplace=True)
it.head(2)
```

```python
fig,ax=plt.subplots(figsize=(5, 5))
it = pd.read_csv(filename)
h2d=plt.hist2d(it['x'], it['y'], bins = (256, 256), range = [(0, 256), (0, 256)])
ax.set_xlim(25,75)
ax.set_ylim(80,110)

n=1
grid_x=np.linspace(25,75,2)
ax.set_xticks(grid_x)
ax.grid(axis="x")

a=np.empty((n,2))
for i in range (1,n+1):
    
    temp=it[(it['x']>grid_x[i-1]) & (it['x'] <grid_x[i])]                          # sub-dataframe for x values between upper and lower bounds of box
    temp_2=temp.groupby(["x",'y']).size().idxmax() 
    a[i-1][0]=temp_2[0]; a[i-1][1]=temp_2[1]

x1=a[0][0]; y1=a[0][1]
print (x1,y1)

fig, (ax0) = plt.subplots(ncols=1, figsize=(2,6))
h = ax0.hist2d(it['x'], it['time'], bins = [256, 400])
#ax0.set_xlim(x1-5,x1+5)
ax0.set_xlabel('x position')
ax0.set_ylabel('Time (ns)')

#it['time'].max()
start = 0; end = it['time'].max(); duration = end - start
ax0.set_ylim(start, end)
```

```python
filename = 'P/x_389.5'
it = pd.read_csv(filename)

filename = 'P/x_389.5_1'
it_1 = pd.read_csv(filename)
```

```python
it_1["time"]+=max(it["time"])
```

```python
meta=pd.concat([it,it_1])
meta.drop("Unnamed: 0",axis=1,inplace=True)
meta.head(2)
```
