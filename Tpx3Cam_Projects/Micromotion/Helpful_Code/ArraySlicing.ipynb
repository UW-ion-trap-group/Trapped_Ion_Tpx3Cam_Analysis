```python
# fit distribution for m bins with double Gauss, fill in arrays for mean and sigma

def GaussianNorm3(x, N1, d): 
    frac = 0.39; sigma1 = 11.7; sigma2 = 26.4; 
#    frac = 0.40; sigma1 = 11.8; sigma2 = 25.8; 
    y = abs(N1)/np.sqrt(2*np.pi)*((1-frac)/sigma1*np.exp(-x**2/(2.*sigma1**2)) + frac/sigma2*np.exp(-x**2/(2.*sigma2**2))) + d 
    return y

ss1 = []; sis1 =[]

for i in range(0, m):
    n1 = 300; offset = 0; ehist =[]
    hist, bin_edges = np.histogram(dtc1bins[i],range=(0, Tlim), bins=nbins) #, density=True)
    
# define Poissonian errors for the histogram
    for jj in range(len(hist)):
        if(hist[jj] != 0): 
            ehistt = np.sqrt(hist[jj])
        else: 
            ehistt = 1. # fitter dont like 0s
        ehist.append(ehistt)
  
    bin_centres = (bin_edges[:-1] + bin_edges[1:])/2
    popt, pcov = curve_fit(GaussianNorm3, bin_centres, hist, [n1, offset], sigma = ehist) 
    hist_fit = GaussianNorm3(bin_centres, *popt)

    plt.figure(figsize=(8, 4))
    plt.plot(bin_centres, hist, 'ro')
    plt.plot(bin_centres, hist_fit)
    plt.show()

    perr = np.sqrt(np.diag(pcov))

    coi = abs(popt[0])*nbins/(2.*Tlim)
    errcoi = abs(perr[0])*nbins/(2.*Tlim)

    print ('i = ', i, 's1 = %7.1f'% coi,'+- %5.1f'% errcoi,' offset = %7.2f'% popt[1],'+- %5.2f'% perr[1])

    ss1.append(coi)
    sis1.append(errcoi)
```
