```python
def Gaussian(x, xm, sigma, A, c):
    return np.absolute(A)*np.exp(-np.power((x-xm)/sigma,2)/2) + c
```

```python
class Ion:
    def __init__(self, n, x, x0, x1, y, y0, y1, r0, data, debugPrinting=False):
        self.n = n
        self.x = x
        self.x0 = x0
        self.x1 = x1
        self.y = y
        self.y0 = y0
        self.y1 = y1
        self.r0 = r0
        self.data = data
        self.debugPrinting = debugPrinting
        
    def save_motion(self, tbins, pbins):
        self.tbins = tbins                                                                                  #Number of time intervals
        self.pbins = pbins                                                                                  #Number of spatial bins 
        
        x = self.data['xc']; y = self.data['yc']; t = self.data['time']                                     #Set up arrays for x,y, and time
        dtbins_x = []; dtbins_y = []                                                                        #create blank bin arrays to be appended
        Tmin = min(data_table['time'])                                                                      #lowest time value of used dataset
        Tmax = max(data_table['time'])                                                                      #Highest time value of used dataset
        
        hist_xt, bins_xt = np.histogram(self.data['xc'], self.pbins, (self.x0,self.x1)) #, density=True)    #X histogram values for integrated time
        hist_yt, bins_yt = np.histogram(self.data['yc'], self.pbins, (self.y0,self.y1)) #, density=True)    #Y histogram values for integrated time
        

        print ('Tmin =', Tmin, ' Tmax = ', Tmax, 'length = ', Tmax-Tmin)                                    #Small check to make sure the code is running

        Tbin = (Tmax - Tmin)/self.tbins                                                                     #Time bin size
        for i in range(0,self.tbins):
            start = Tmin + i*Tbin; stop = start + Tbin                                                      #beginning and end times for each time interval
            dtbins_x.append([])                                                                             #create blank array to fill later, within original array
            dtbins_y.append([])                                                                             #create blank array again for y
            if (i/10).is_integer() is True:
                print(i)                                                                                    #Another check to make sure cod is running
            for j in range(len(x)):
                if(start < t[j] < stop): 
                    dtbins_x[i].append(x[j])
                    dtbins_y[i].append(y[j])
            
        mean_x = []; mean_y = []
        sigma_x = []; sigma_y = []
        amp_x = []; amp_y = []
        c_x = []; c_y = []
        
    def save_micromotion(self, tbins, pbins):
        self.tbins = tbins                                                                                     #Number of time intervals
        self.pbins = pbins                                                                                     #Number of spatial bins
        
        x = self.data['xc']; y = self.data['yc']; t = self.data['tfold']                                       #Set up callable arrays for x,y, and time
        dtbins_x = []; dtbins_y = []                                                                           #create blank bin arrays to be appended
        Tmin = min(data_table['tfold'])                                                                        #lowest time value of used dataset
        Tmax = max(data_table['tfold'])                                                                        #Highest time value of used dataset
        
        hist_xt, bins_xt = np.histogram(self.data['xc'], self.pbins, (self.x-np.sqrt(2),self.x+np.sqrt(2))) #, density=True)   #X histogram values for integrated time
        hist_yt, bins_yt = np.histogram(self.data['yc'], self.pbins, (self.y-np.sqrt(2),self.y+np.sqrt(2))) #, density=True)   #Y histogram values for integrated time
        

        print ('Tmin =', Tmin, ' Tmax = ', Tmax, 'length = ', Tmax-Tmin)                                       #Small check to make sure the code is running

        Tbin = (Tmax - Tmin)/self.tbins                                                                        #Time bin size
        for i in range(0,self.tbins):
            start = Tmin + i*Tbin; stop = start + Tbin                                                         #beginning and end times for each time interval
            dtbins_x.append([])                                                                                #create blank array to fill later, within original array
            dtbins_y.append([])                                                                                #create blank array again for y
            if (i/10).is_integer() is True:
                print(i)                                                                                       #Another check to make sure code is running
            for j in range(len(x)):
                if(start < t[j] < stop):                                                                       #Only use data that is within the time interval
                    dtbins_x[i].append(x[j])                                                                   #Append x values
                    dtbins_y[i].append(y[j])                                                                   #Append y values
            
        mean_x = []; mean_y = []                                                                               #Create arrays that will hold Gaussian fit data
        sigma_x = []; sigma_y = []                                                                             #For mean, sigma, amplitude, and base fluorescence
        amp_x = []; amp_y = []
        c_x = []; c_y = []
        
        xguess_mean = (self.x)                                                                                 #These are all of the fit guesses
        xguess_sigma = 1
        xguess_amp = 700/tbins
        xguess_c = 700/tbins/10
        guess_x  = np.array([xguess_mean,xguess_sigma,xguess_amp,xguess_c])                                     #X initial guess array
            
        yguess_mean = (self.y)                                                                                 #Fit guesses for Y
        yguess_sigma = 1
        yguess_amp = 700/tbins
        yguess_c = 700/tbins/10
        guess_y  = np.array([yguess_mean,yguess_sigma,yguess_amp,yguess_c])                                     #Y initial guess array
        
               
        for i in range(0, self.tbins):                                                                                     #Call data for each time interval
            hist_x, bins_x = np.histogram(dtbins_x[i], 8, (self.x-np.sqrt(2),self.x+np.sqrt(2))) #, density=True),                           #x histogram info 
            bin_centres_x = (bins_x[:-1] + bins_x[1:])/2                                                                   #data centers not edges
            
            hist_y, bins_y = np.histogram(dtbins_y[i], 11, (self.y-np.sqrt(2),self.y+np.sqrt(2))) #, density=True)                           #Same for Y
            bin_centres_y = (bins_y[:-1] + bins_y[1:])/2
            
            popt_x, pcov_x = scipy.optimize.curve_fit(Gaussian, bin_centres_x, hist_x, p0=guess_x, maxfev = 50000)          #using the guesses, create a Gaussian
            mean_x.append(popt_x[0]); sigma_x.append(np.sqrt(np.abs(pcov_x[0,0]))); amp_x.append(popt_x[2]); c_x.append(popt_x[3])    #append data to the arrays
            hist_fit_x = Gaussian(bin_centres_x, *popt_x)                                                                  #only for plotting
            
            popt_y, pcov_y = scipy.optimize.curve_fit(Gaussian, bin_centres_y, hist_y, p0=guess_y, maxfev = 50000)          #use guesses to create a gaussian
            mean_y.append(popt_y[0]); sigma_y.append(np.sqrt(np.abs(pcov_y[0,0]))); amp_y.append(popt_y[2]); c_x.append(popt_y[3])    #append y data to arrays
            hist_fit_y = Gaussian(bin_centres_y, *popt_y)
            
        ion_micromotion_data[f'x{self.n}'] = mean_x                                                                        #save all fit data into a permanent table that can be called. 
        ion_micromotion_data[f'y{self.n}'] = mean_y
        ion_micromotion_data[f'sigma_x{self.n}'] = sigma_x
        ion_micromotion_data[f'sigma_y{self.n}'] = sigma_y
        ion_micromotion_data[f'amp_x{self.n}'] = amp_x
        ion_micromotion_data[f'amp_y{self.n}'] = amp_y
            
        #fig, ax = plt.subplots(3, 2, figsize=(10, 15))
        #ax[0,0].plot(mean_x); ax[0,0].set_title(f"{self.n}X Mean");
        #ax[0,1].plot(mean_y); ax[0,1].set_title(f"{self.n}Y Mean")
        #ax[1,0].plot(sigma_x); ax[1,0].set_title(f"{self.n}X Sigma"); #ax[1,0].set_ylim(-5,7)
        #ax[1,1].plot(sigma_y);ax[1,1].set_title(f"{self.n}Y Sigma"); #ax[1,1].set_ylim(-2,5)
        #ax[2,0].plot(amp_x); ax[2,0].set_title(f"{self.n}X Amplitude"); #ax[2,0].set_ylim(top=300)
        #ax[2,1].plot(amp_y); ax[2,1].set_title(f"{self.n}Y Amplitude"); #ax[2,1].set_ylim(top=260)
        
        #fig = plt.figure(figsize=(5,5))
        #ax = plt.axes(projection='3d')
        #ax.plot3D(mean_x, mean_y, tbins, label='ion path')
        #ax.legend()
        #fig, ax = plt.subplots(1, 1, figsize=(5, 5))
        #ax.plot(mean_x, mean_y, label=f'Ion {self.n} path')
        
    def micromotion_four_styles(self):
        pbins = 8
        tbins = 100      
        n=0
        used_data = self.data
        t = used_data['tfold']
        Tmin = t.min()
        Tmax = t.max()
        mean_x = self.x; mean_y = self.y
        sigma_x = 1; sigma_y = 1
        amp_x = 500; amp_y = 500
        c_x = 0; c_y = 0
        xguess  = np.array([mean_x,sigma_x,amp_x,c_x])
        yguess  = np.array([mean_y,sigma_y,amp_y,c_y])

        xgrid = np.linspace(self.x-2, self.x+2, pbins)
        ygrid = np.linspace(self.y-2, self.y+2, pbins)
        Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
            
        print ('Tmin =', Tmin, ' Tmax = ', Tmax, 'length = ', Tmax-Tmin)

        Tbin = (Tmax - Tmin)/tbins
        for i in range(0,tbins):
            if (i/10).is_integer() is True:
                print(i)
            start = Tmin + i*Tbin; stop = start + Tbin
            used = (self.data
                    .query(f"{start}< tfold <{stop}"))
    
            
        
            hist_x, bins_x = np.histogram(used['xc'], pbins, ) #, density=True),
            bin_centres_x = (bins_x[:-1] + bins_x[1:])/2
            
            hist_y, bins_y = np.histogram(used['yc'], pbins, ) #, density=True) 
            bin_centres_y = (bins_y[:-1] + bins_y[1:])/2
                            
            popt_x, pcov_x = scipy.optimize.curve_fit(Gaussian, bin_centres_x, hist_x, p0=xguess, maxfev = 50000)
            hist_fit_x = Gaussian(bin_centres_x, *popt_x)
            
            popt_y, pcov_y = scipy.optimize.curve_fit(Gaussian, bin_centres_y, hist_y, p0=yguess, maxfev = 50000)
            hist_fit_y = Gaussian(bin_centres_y, *popt_y)
            mean_x = popt_x[0]; mean_y = popt_y[0]
            sigma_x = popt_x[1]; sigma_y = popt_y[1]
            amp_x = popt_x[2]; amp_y = popt_y[2]
            c_x = popt_x[3]; c_y = popt_y[3]
    
            data = np.vstack([used['xc'], used['yc']])
            kde = gaussian_kde(data)
            Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))
    
    
            fig, ([[ax1, ax2], [ax3, ax4]]) = plt.subplots(ncols=2, nrows=2, figsize=(11,11))
            pcm = sns.kdeplot(used['xc'], used['yc'], shade=True, n_levels=30, ax=ax1)
            ax1.plot(mean_x, mean_y, 'ro', markersize=10)
            ax1.plot(mean_x, mean_y, 'r+', markersize=50)
            ax2.imshow(Z.reshape(Xgrid.shape),
                   origin='lower', aspect='auto',
                   extent=[self.x-2, self.x+2, self.y-2, self.y+2],
                   cmap='viridis', vmin=0, vmax=.12)
            ax2.plot(mean_x, mean_y, 'ro', markersize=10)
            ax2.plot(mean_x, mean_y, 'r+', markersize=50)
            ax3.hist2d(used['xc'], used['yc'], bins=16, range = [(self.x-2,self.x+2), (self.y-2, self.y+2)])
            ax3.plot(mean_x, mean_y, 'ro', markersize=10)
            ax3.plot(mean_x, mean_y, 'r+', markersize=50)
            ax4.hist2d(used['x'], used['y'], bins=4, range = [(self.x-2,self.x+2), (self.y-2, self.y+2)])
            ax4.plot(mean_x, mean_y, 'ro', markersize=10)
            ax4.plot(mean_x, mean_y, 'r+', markersize=50)
            ax1.set_xlim(self.x-2,self.x+2)
            ax1.set_ylim(self.y-2,self.y+2)
            ax2.set_xlim(self.x-2,self.x+2)
            ax2.set_ylim(self.y-2,self.y+2)
            ax3.set_xlim(self.x-2,self.x+2)
            ax3.set_ylim(self.y-2,self.y+2)
            ax4.set_xlim(self.x-2,self.x+2)
            ax4.set_ylim(self.y-2,self.y+2)
            fig.savefig(f'Micromotion {self.n}/{n}.png')
            plt.close(fig)
            n=n+1

        images = [] 
        n=0
        for i in range(0,tbins):
            h = Image.open(f'Micromotion {self.n}/{n}.png')
            images.append(h)
            n=n+1
    
        images[0].save(f'Micromotion GIFs/Micromotion {self.n}.gif',
                       save_all=True,
                       append_images=images[1:],
                       duration=100,
                       loop=0)
        
    def micromotion_track(self):
        pbins = 8
        tbins = 100      
        n=0
        used_data = self.data
        t = used_data['tfold']
        Tmin = t.min()
        Tmax = t.max()
        mean_x = self.x; mean_y = self.y
        sigma_x = 1; sigma_y = 1
        amp_x = 500; amp_y = 500
        c_x = 0; c_y = 0
        xguess  = np.array([mean_x,sigma_x,amp_x,c_x])
        yguess  = np.array([mean_y,sigma_y,amp_y,c_y])
        x = []
        y = []
            
        print ('Tmin =', Tmin, ' Tmax = ', Tmax, 'length = ', Tmax-Tmin)

        Tbin = (Tmax - Tmin)/tbins
        for i in range(0,tbins):
            if (i/10).is_integer() is True:
                print(i)
            start = Tmin + i*Tbin; stop = start + Tbin
            used = (self.data
                    .query(f"{start}< tfold <{stop}"))
    
            
        
            hist_x, bins_x = np.histogram(used['xc'], pbins, (self.x-2,self.x+2)) #, density=True),
            bin_centres_x = (bins_x[:-1] + bins_x[1:])/2
            
            hist_y, bins_y = np.histogram(used['yc'], pbins, (self.y-2,self.y+2)) #, density=True) 
            bin_centres_y = (bins_y[:-1] + bins_y[1:])/2
                            
            popt_x, pcov_x = scipy.optimize.curve_fit(Gaussian, bin_centres_x, hist_x, p0=xguess, maxfev = 50000)
            hist_fit_x = Gaussian(bin_centres_x, *popt_x)
            
            popt_y, pcov_y = scipy.optimize.curve_fit(Gaussian, bin_centres_y, hist_y, p0=yguess, maxfev = 50000)
            hist_fit_y = Gaussian(bin_centres_y, *popt_y)
            mean_x = popt_x[0]; mean_y = popt_y[0]
            sigma_x = popt_x[1]; sigma_y = popt_y[1]
            amp_x = popt_x[2]; amp_y = popt_y[2]
            c_x = popt_x[3]; c_y = popt_y[3]  
            
            perr_x = np.sqrt(np.abs(np.diag(pcov_x)))
            perr_y = np.sqrt(np.abs(np.diag(pcov_y)))
    
            fig, (ax1) = plt.subplots(ncols=1, figsize=(3,3))
            ax1.hist2d(used['xc'], used['yc'], bins=pbins, range = [(self.x-2,self.x+2), (self.y-2, self.y+2)])
            if self.x-.5 < mean_x < self.x+.5 and self.y-1 < mean_y < self.y+1:
                x.append(mean_x); y.append(mean_y)
            ax1.plot(x, y, 'b', linewidth=1)
            if self.x-.5 < mean_x < self.x+.5 and self.y-1 < mean_y < self.y+1 and perr_x[0] < 1 and perr_y[0] < 1:
                ax1.errorbar(mean_x, mean_y, perr_x[0], perr_y[0], 'ro', markersize=3, ecolor='r')
            ax1.set_xlim(self.x-2,self.x+2)
            ax1.set_ylim(self.y-2,self.y+2)

            fig.savefig(f'Micromotion {self.n}/{n}.png')
            plt.close(fig)
            n=n+1

        images = [] 
        n=0
        for i in range(0,tbins):
            h = Image.open(f'Micromotion {self.n}/{n}.png')
            images.append(h)
            n=n+1
    
        images[0].save(f'Micromotion GIFs/Micromotion Tracked {self.n}.gif',
                       save_all=True,
                       append_images=images[1:],
                       duration=100,
                       loop=0)
```
