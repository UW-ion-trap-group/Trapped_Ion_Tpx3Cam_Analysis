# Combine Datasets

*****This type of analysis has been seen to act differently when applying the threshold setting method and may lead to data that is not representative of the science...*****

This Notebook serves to combine **pandas** dataframes that have previously been constructed in the **Jumps_Data_Pandas.ipynb** Notebook. 

Specific dataframes are called and operated on by using the **choose_file.____** command which performs a function within the module **choose_file.py** that pulls data from a specific file and assigns variables based on the information in that file. 

The **append_file(n):** function defined near the top is the function that allows for the combination of information from multiple dataframes to be combined and later analyzed. Before performing an **append_file(n)** command, the file from which you want to append the data must be selected by using the **choose_file.____** command. 

The functions from **File_Functions.py** are used to perform analysis that combines data from all of the ions in the set and generalizes the behavior of the ions in that set. When performed after a number of **append_file(n)** functions, this will generalize the data for all of the ions in all of the sets that have been combined. 

```python
### All of these imports ar not necessary, but I have
### included them because they've been used in previous
### notebooks and I don't actually know which are used in this

import numpy as np
import scipy
import scipy.optimize
from scipy.optimize import curve_fit
from scipy import stats
import scipy.stats
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.colors as mcolors
from mpl_toolkits import mplot3d
import glob
import time
import pandas as pd
import math
plt.rcParams["figure.figsize"] = (3,3)

# Module must be in the same folder as this Notebook
import File_functions



print (time.ctime())
```

```python
# Modules must be in the same folder as this Notebook
import Ion_functions as func
from Ion_functions import Ion
import choose_file 

# Function that allows for the newly called file information to be combined to 
# the 'global' information in this Notebook. Essentially, this allows for us to 
# combine pandas dataframes into a single dataframe for analysis. 
def append_file(n):
    # Allows writing to the the global variables of the Notebook. 
    global Ion_1; global Ion_2; global Ion_3; 
    global Ion_4; global Ion_5; global Ion_6; global time
    
    # Append new data to the current dataframe. Only change is that all of the new data 
    # timestamps have been added with the last time stamp of the current dataframe. This
    # makes sure to keep timing non-degenerate and continuous as if it were a single file
    
    # Subtracting 1 from n each time will stop the function after the proper number of ions
    # have been appended. 
    
    if n > 0:   
        time += choose_file.data_table['time'].max()
        Ion_1_to_append = choose_file.Ion_1
        Ion_1_to_append['time'] = Ion_1_to_append['time'] + Ion_1['time'].max()
        Ion_1 = Ion_1.append(Ion_1_to_append)
        n = int(n-1)
        
    if n > 0: 
        Ion_2_to_append = choose_file.Ion_2
        Ion_2_to_append['time'] = Ion_2_to_append['time'] + Ion_2['time'].max()
        Ion_2 = Ion_2.append(Ion_1_to_append)
        n = int(n-1)
        
    if n > 0:    
        Ion_3_to_append = choose_file.Ion_3
        Ion_3_to_append['time'] = Ion_3_to_append['time'] + Ion_3['time'].max()
        Ion_3 = Ion_3.append(Ion_1_to_append)
        n = int(n-1)
    
    if n > 0:     
        Ion_4_to_append = choose_file.Ion_4
        Ion_4_to_append['time'] = Ion_4_to_append['time'] + Ion_4['time'].max()
        Ion_4 = Ion_4.append(Ion_1_to_append)
        n = int(n-1)
    
    if n > 0:     
        Ion_5_to_append = choose_file.Ion_5
        Ion_5_to_append['time'] = Ion_5_to_append['time'] + Ion_5['time'].max()
        Ion_5 = Ion_5.append(Ion_1_to_append)
        n = int(n-1)
    
    if n > 0:     
        Ion_6_to_append = choose_file.Ion_6
        Ion_6_to_append['time'] = Ion_6_to_append['time'] + Ion_6['time'].max()
        Ion_6 = Ion_6.append(Ion_1_to_append)
```
