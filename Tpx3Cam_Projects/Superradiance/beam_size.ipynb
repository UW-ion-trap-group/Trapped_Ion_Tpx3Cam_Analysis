```python
choose_file.yscan_149s() #file selection, plots ROI's to verify number and location of Ions. 
ion_1 = choose_file.ion_1
ion_2 = choose_file.ion_2
ion_3 = choose_file.ion_3
ion_4 = choose_file.ion_4
Ion_1 = choose_file.Ion_1
Ion_2 = choose_file.Ion_2
Ion_3 = choose_file.Ion_3
Ion_4 = choose_file.Ion_4



filename = choose_file.filename
old_data_table = choose_file.old_data_table
data_table = choose_file.data_table
time = data_table['time'].max() - data_table['time'].min()


sigma = 2
uncertainty = False
single_photon = False
ion_1.setup(sigma, uncertainty, single_photon)
ion_2.setup(sigma, uncertainty, single_photon)
ion_3.setup(sigma, uncertainty, single_photon)
ion_4.setup(sigma, uncertainty, single_photon)

rates_yscan_149s = [len(ion_1.transpts)/time, len(ion_2.transpts)/time, len(ion_3.transpts)/time, len(ion_4.transpts)/time]
positions_yscan_149s = [ion_1.x, ion_2.x, ion_3.x, ion_4.x]
xs = [ion_1.x, ion_2.x, ion_3.x, ion_4.x]
ys = [ion_1.y, ion_2.y, ion_3.y, ion_4.y]

separations_yscan_149s = []
for i in range(len(xs)): 
    it = []
    for j in range(len(xs)):
        it.append(np.sqrt((xs[i] - xs[j])**2 + (ys[i] - ys[j])**2))
    separations_yscan_149s.append(np.average(it))
    
fig, ax = plt.subplots(1, figsize = (3,3))
ax.scatter(positions_yscan_149s, rates_yscan_149s)
```
